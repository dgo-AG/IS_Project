```python
#Used to import both pandas and maplotlib in order to create and use the graphs and charts displayed in the following code.
import matplotlib.pyplot as plt
import pandas as pd
```

# __*1.*  What is the average probability rate for all jobs to be automated by AI by the year 2030?__


```python
# The csv data file is loaded in
df = pd.read_csv("AI_Impact_on_Jobs_2030.csv")

# The average probability rate of the jobs is calculated
avg_prob_rate = df['Automation_Probability_2030'].mean()

# The histogram is plotted and formatted
plt.figure(figsize=(8, 4))
plt.hist(df['Automation_Probability_2030'], edgecolor='black', color='indigo')
plt.title('AI Automation Probability Distribution by Year 2030', fontsize=15)
plt.xlabel('Rate of Automation Probability', fontsize=10)
plt.ylabel('Amount of Jobs', fontsize=10)

# A vertical dotted line to show the average probability overall is created
plt.axvline(avg_prob_rate, color = 'skyblue', linestyle = 'dotted', linewidth = 1.5, label = f'Probability Average: {avg_prob_rate:.2f}')
plt.grid(axis = 'y', alpha = .70)
plt.legend()

# The calculated average is displayed with the print function and the format is made to fit
plt.tight_layout()
print(f"The exact average automation probability rate for the jobs is: {avg_prob_rate}")
```

# __*2.*  Is there a correlation between the education level required for a job and its probability to be automated by AI in 2030?__


```python
df = pd.read_csv("AI_Impact_on_Jobs_2030.csv")

# A group is created with Education_Level and calculates the average with Automation_Probability_2030
avg_prob_edu = df.groupby('Education_Level')['Automation_Probability_2030'].mean().sort_values(ascending=False)

# A bar chart is created and formatted using matplotlib
plt.figure(figsize=(10, 5))
plt.bar(avg_prob_edu.index, avg_prob_edu.values, color=['cyan', 'navajowhite', 'seagreen', 'mediumorchid'])
plt.xlabel('Level of Education', fontsize=10)
plt.ylabel('Average Probability for Automation by 2030', fontsize=10)
plt.title('Average Automation Probability Based on Education Level', fontsize=15)
plt.xticks(rotation=50)
plt.grid(axis='y', linestyle='dotted', alpha=.70)

#The answer to if there's a correlation is answered with print and the format is made to fit
plt.tight_layout()
print("There is no kind of correlation between the two.")
```
