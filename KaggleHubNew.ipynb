## Question 5
What proportion of jobs offer high salaries and require advanced AI skills?


```python
import pandas as pd # importing pandas
import matplotlib.pyplot as plt # importing matplotlib

df = pd.read_csv("AI_Impact_on_Jobs_2030.csv") # importing the data set


high = len(df[(df["Average_Salary"] >= 50000) & (df["Skill_1"] > 0.5)]) # Counting how many jobs have an average salary of $50,000 or more 
# AND have a Skill_1 value greater than 0.5
low = len(df) - high # Count the remaining jobs (those that donâ€™t meet the conditions above)


plt.pie([high, low], labels=["High Salary & Skill_1 > 0.5", "Other Jobs"], autopct='%1.1f%%') # Created a pie chart showing the two groups 
# which contained High salary & Skill_1 > 0.5 and All other jobs
plt.show() # show me the pie chart
```

## Question 6

Which skills drive higher salaries and are most valuable for career growth?

```python
df = pd.read_csv("AI_Impact_on_Jobs_2030.csv") # importing the data set


skills = ["Skill_1", "Skill_2", "Skill_3"]  # Creating a list of the skill columns I wanted on the chart
corr = df[skills + ["Average_Salary"]].corr()["Average_Salary"].drop("Average_Salary") # shows how each skill is correlated with the salaries

corr.plot(kind="bar", color="teal") #  Made a bar chart of the correlations on the image below
plt.ylabel("Correlation with Salary") # giving the y value a name
plt.show() # show the chart
```

## Question 7

How many jobs require strong AI skills and have high demand?

```python
df = pd.read_csv("AI_Impact_on_Jobs_2030.csv") # importing the data set

high = len(df[df["Skill_1"] > 0.5]) # Count jobs where Skill_1 is above 0.5
low = len(df) - high # Count the remaining jobs

plt.pie([high, low], labels=["High AI Skill", "Other Jobs"], autopct='%1.1f%%') # Created a pie chart showing both groups
plt.show() # Show the pie chart
```

## Question 8

Which AI skill and education combinations correspond to the highest-paying jobs?


```python
df = pd.read_csv("AI_Impact_on_Jobs_2030.csv") # load up the dataset

df['Skill_Edu'] = df['Skill_1'].round(1).astype(str) + " Skill | " + df['Education_Level'] # Create a new 
# column combining Skill_1 values with education level

top_combos = df.groupby('Skill_Edu')['Average_Salary'].mean().sort_values(ascending=False).head(10)
#  Finding the top 10 skills + education combinations with the highest average salary

plt.xlabel("Average Salary") # label x axis
plt.ylabel("AI Skill + Education") # label y axis
plt.gca().invert_yaxis() # Highest salaries appear at the top 
plt.show()  # Displays the chart
```
